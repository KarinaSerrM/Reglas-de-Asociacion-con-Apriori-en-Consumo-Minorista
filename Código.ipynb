```python
# Importamos las posibles librerias a utilizar en caso de requerir otra en el transcurso la agregamos
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Creamos un DataFrame a partir de los datos proporcionados en el ejercicion, agregando una columna de ticket para tenerlo como referencia
data = {'TicketNo': ['Ticket_1', 'Ticket_2', 'Ticket_3', 'Ticket_4', 'Ticket_5', 
                 'Ticket_6', 'Ticket_7', 'Ticket_8', 'Ticket_9', 'Ticket_10', 'Ticket_11'],
    'Products': [['bread', 'butter', 'wine', 'bananas', 'coffee', 'carrots'],
        ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
        ['beer', 'chips', 'asparagus', 'salsa', 'milk', 'apples'],
        ['olive oil', 'bread', 'butter', 'tomatoes', 'steak', 'carrots'],
        ['tomatoes', 'onions', 'chips', 'wine', 'ketchup', 'oranges'],
        ['bread', 'butter', 'beer', 'chips', 'milk'],
        ['butter', 'tomatoes', 'carrots', 'coffee', 'sugar'],
        ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
        ['bread', 'butter', 'ketchup', 'coffee', 'chicken wings'],
        ['butter', 'beer', 'chips', 'asparagus', 'apples'],
        ['tomatoes', 'onion', 'beer', 'chips', 'milk', 'coffee']]}
my_basket = pd.DataFrame(data)
my_basket
```

```python
# Ahora, vamos a crear una copia para expandir los productos 
expanded = my_basket.explode('Products')
# Consolidar por TicketNo y producto (contando ocurrencias)
basket = (expanded.groupby(['TicketNo', 'Products']).size().unstack(fill_value=0).reset_index().set_index('TicketNo'))
basket
```

```python
# Aplicamos el algoritmo "a priori" para localizar los artículos frecuentemente comprados en la base de datos
from mlxtend.frequent_patterns import apriori
```

```python
# analisis "a priori" con un soporte mínimo del 25% tomo este criterio porque nuestra basde de datos en muy pequeña
articulos_frecuentes = apriori(basket, min_support = 0.25, use_colnames = True).sort_values('support', ascending = False).reset_index(drop = True)
articulos_frecuentes['length'] = articulos_frecuentes['itemsets'].apply(lambda x:len(x))
articulos_frecuentes
```

Vemos los articulos y el soporte en función de la frecuencia que llegamos a encontrar, eEsto es preeliminar tenemos un total de82 diferentes articulos
• Ahora vamos a hacer las reglas de asociación:

```python
 ## Definimos las reglas de asociación
association_rules(articulos_frecuentes, metric = 'confidence',
                 min_threshold = 0.50).sort_values('confidence', ascending = False).reset_index(drop=True)
```

```python
# Definimos las reglas de asociación
association_rules(articulos_frecuentes, metric = 'lift',
                 min_threshold = 1.01).sort_values('lift', ascending = False).reset_index(drop=True)
```

## Resumen del Análisis de las Reglas de Asociación (confidence).
Para este ejercicio las reglas las filtré con un umbral de confianza (`min_threshold`) de 0.5, lo que nos garantiza que solo se consideren relaciones significativas entre productos.
#### **Principales Patrones de Consumo**
Los patrones en las reglas de los clientes son los siguientes:

- **Combinaciones Fuertemente Relacionadas**:
- Los clientes que compran `bread` siempre compran `butter`
- Los clientes que compran `beer` tienen una alta probabilidad de comprar `chips`esto en el  75% de los casos
- Los clientes que compran `carrots` tienden a comprar `butter` en el **80%** de los casos
- La combinación de `chips` y `milk` está fuertemente asociada con la compra de `beer` igual en un 75% de las veces

- **Productos Individuales Populares**:
Por otro lado tenemos productos individuales que son muy populares, como por ejemplo: 
- **tomatoes**: Aparece como consecuencia en varias reglas (por ejemplo, `onions -> tomatoes`).   
- **milk**: También es un producto popular, apareciendo tanto como antecedente como consecuente en varias reglas. Su alta frecuencia indica que es un artículo esencial en muchas transacciones.
  
#### **Oportunidades de Marketing**
Basándose en estas reglas, se pueden implementar estrategias de marketing efectivas, como lo son las **Promociones Cruzadas**, **Colocación Estratégica**, **Recomendaciones Personalizadas** e incluso la **Segmentación de Clientes**


### Resumen del Análisis de las Reglas de Asociación (lift)

Para este caso, las reglas han sido filtradas con un umbral de **lift** (`min_threshold`) de 1.01, lo que garantiza que solo se consideren relaciones significativas entre productos.

#### **Principales Patrones de Consumo**
Los patrones que se observan son los siguientes:

- **Combinaciones Fuertemente Relacionadas (Altos Valores de Lift)**:
  
- Los clientes que compran `beer` tienden a comprar `chips` y `milk` juntos, lo que sugiere una relación fuerte entre estos productos.
- La combinación de `chips` y `milk` está altamente asociada con la compra de `beer`.
- La relación entre `tomatoes` y `onions` es significativa, ya que la compra de uno incrementa considerablemente la probabilidad de comprar el otro.
- Los clientes que compran `butter` tienen una alta probabilidad de comprar `carrots`.

- **Productos Individuales Populares**:
- **`tomatoes` y `onions`**: Ambos productos aparecen en varias reglas, mostrando una relación estrecha. Por ejemplo, `onions -> tomatoes` tiene un **lift de 1.45**, lo que indica que la compra de `onions` incrementa significativamente la probabilidad de comprar `tomatoes`.
- **`butter` y `carrots`**: La regla `butter -> carrots` tiene un **lift de 1.83**, lo que sugiere que los clientes que compran `butter` tienden a comprar `carrots` con mayor frecuencia de lo esperado por casualidad.
